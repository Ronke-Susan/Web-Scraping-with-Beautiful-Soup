**Web Scraping Assignment**

Scrape the lyrics of the poem "99 bottles of beer on the wall" from 'https://www.songfacts.com/lyrics/traditional/99-bottles-of-beer'

```python
# importing libraries
from bs4 import BeautifulSoup
import requests
```

```python
# Assigning link to a variable
url = 'https://www.songfacts.com/lyrics/traditional/99-bottles-of-beer'

# using request to get the raw HTML content from the website
HTML_script = requests.get(url)
```

```python
HTML_script
```

<Response [200]> implies the request was successful

```python
# Extracting the text from the HTML_script
text_script = HTML_script.text
```

```python
text_script
```

```python
useful_text = BeautifulSoup(text_script, "html.parser") 
#printing the useful_text using prettify module to make it readable 
print(useful_text.prettify())
```

The HTML parser is a structured markup processing tool. It defines a class called HTMLParser, which is used to parse HTML files

```python
useful_text.find_all('div', class_="inner")
#finds all content under the tag name div and class specified
```

```python
#cleaning the scraped data
def pretty_text(text):
    '''remove white spaces and unwanted characters'''
    final = (((text).replace(u'<br/>',' ').replace(u'\n',' ').lstrip().rstrip()))
    
    return final
poem = (pretty_text(useful_text.find_all('div', class_="inner")[0].text))
```
